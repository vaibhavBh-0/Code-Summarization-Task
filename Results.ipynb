```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('history_df.csv')
```

```python
epochs = df['Epochs']
```

```python
df
```

```python
acc_features = ['Training Accuracy', 'Validation Accuracy']
loss_features = ['Training Loss', 'Validation Loss']
```

```python
for feature in acc_features + loss_features:
    df[feature] = df[feature].apply(lambda x: float(x.split('(')[1].split(',')[0]))
```

```python
df
```

```python
plt.figure(figsize=(10, 8))

fig = plt.plot(epochs + 1, df[acc_features], label=acc_features)
_ = plt.xlabel('Epochs')
_ = plt.ylabel('Accuracy')
_ = plt.title('Training and Validation Accuracy')
_ = plt.legend()

plt.savefig('results/accuracy.png', transparent=False)
```

```python
plt.figure(figsize=(10, 8))

_ = plt.plot(epochs + 1, df[loss_features], label=loss_features)
_ = plt.xlabel('Epochs')
_ = plt.ylabel('Loss value')
_ = plt.title('Training and Validation Loss')
_ = plt.legend()

plt.savefig('results/loss.png', transparent=False)
```
